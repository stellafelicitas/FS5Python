**[Task 1.2 (a).]()** Use `plt.imread` to load an image:

**[Task 1.2 (d).]()** Binarize the image by applying an intensity threshold:

```python
T = 0.1 
binary_image = (image > T).astype(int)

plt.imshow(binary_image, cmap='gray')
plt.show()
```

**[Task 1.2 (e).]()** Load the ground truth:

```python
gt = plt.imread('data/NIH3T3/gt/0.png')
```

Compute and print the Dice coefficient:

```python
def dice_coefficient(binary_image, gt):
    intersection = np.sum(binary_image * gt) # |G  H| is the intersection (shared pixels between binary_image and gt)
    # |G| + |H| is the sum of pixels in both images
    # Multiplying by 2 ensures the result is normalized between 0 and 1
    return 2 * intersection / (np.sum(binary_image) + np.sum(gt))

dice_score = dice_coefficient(binary_image, gt)
print("Dice-Koeffizient:", dice_score)
```
