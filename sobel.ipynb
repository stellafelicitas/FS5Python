**[Task 2.2.]()** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical $3 \times 3$ Sobel filtering, respectively:

```python
sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]) / 8 
sobel_y = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]]) / 8
[rows, columns] = np.shape(img)  # we need to know the shape of the input grayscale image
sobel_img = np.zeros(shape=(rows, columns))  # initialization of the output image array (all elements are 0)
```

```python
gx = np.zeros((rows, columns))
gy = np.zeros((rows, columns))
mag = np.zeros((rows, columns))

for i in range(rows - 2):
    for j in range(columns - 2):
        gx[i + 1, j + 1] = np.sum(np.multiply(sobel_x, img[i:i + 3, j:j + 3]))  # x-Richtung
        gy[i + 1, j + 1] = np.sum(np.multiply(sobel_y, img[i:i + 3, j:j + 3]))  # y-Richtung
        mag[i + 1, j + 1] = np.sqrt(gx[i + 1, j + 1] ** 2 + gy[i + 1, j + 1] ** 2)  # Magnitude
```

**[Task 2.3.]()** Test your implementations from above:

```python
# Ergebnisse visualisieren
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(gx, cmap='gray')
axes[0].set_title("gx (Gradient in x-Richtung)")
axes[0].axis('off')

axes[1].imshow(gy, cmap='gray')
axes[1].set_title("gy (Gradient in y-Richtung)")
axes[1].axis('off')

axes[2].imshow(mag, cmap='gray')
axes[2].set_title("Magnitude (gx + gy)")
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
